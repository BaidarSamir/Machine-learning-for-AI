```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import random

plt.style.use("ggplot")
random.seed(3)
samples=40
year_expere=np.random.randint(1,11,samples)
basesalary=2000
randombonus=np.random.randint(1,401,samples)
sal=basesalary+200*year_expere+randombonus

data={
    'years': year_expere,
    'salary':sal
}
df= pd.DataFrame(data)

print(df)
df.to_csv('dataframeforsalaire.csv')
xticks_used=np.arange(0,11)
years = np.arange(1, 11, 1)  
yticks_used=np.linspace(2000,4500,11)
year_exper=np.array([[10]])
data=pd.read_csv('dataframeforsalaire.csv')
year=data["years"]
salaire=data["salary"]
X = data['years'].values.reshape(-1, 1)
y = data['salary'].values

model = LinearRegression()
model.fit(X, y)
plt.plot(X, model.predict(X), color='red', linewidth=2, label='Regression line')
print(model.predict(year_exper))


plt.title("Salaire de chaque employee")
plt.scatter(year,salaire, label="scatter plot of the salaries",alpha=0.7,s=100,c=y)
plt.xticks(xticks_used)
plt.yticks(yticks_used)
plt.xlabel("years of experience")
plt.ylabel("salary")
plt.legend()
plt.show()




```
