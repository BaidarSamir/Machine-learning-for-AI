```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


plt.style.use("fivethirtyeight")
data=pd.read_csv("data (3).csv")

data.head(5)

```

```python
bear_counts = data['Type of bear'].value_counts()


plt.figure(figsize=(8, 6))
bear_counts.plot(kind='bar', color='#25e9f5')


plt.title('Count of Different Types of Bears Involved in Incidents', fontsize=14)
plt.xlabel('Type of Bear', fontsize=12)
plt.ylabel('Count', fontsize=12)


plt.xticks(rotation=45, )
plt.show()
```

```python

age_killed = data[data['Only one killed'] == 1][' age'].dropna()


plt.figure(figsize=(8, 6))
plt.hist(age_killed, bins=10, color='salmon', edgecolor='black')


plt.title('Distribution of Ages of People Who Died in Bear Incidents', fontsize=14)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Frequency', fontsize=12)


plt.show()
```
