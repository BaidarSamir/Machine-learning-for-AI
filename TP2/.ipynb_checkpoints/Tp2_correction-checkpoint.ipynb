```python
X_Age=X.drop(columns="sex")
X_Age.head(5)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


X_Age_encoded = pd.get_dummies(X_Age, drop_first=True)  

correlation_matrix = X_Age_encoded.corr()


plt.figure(figsize=(12, 12))
heatmap = sns.heatmap(correlation_matrix, cmap='coolwarm', vmin=-1, vmax=1, annot=True, fmt=".2f")

plt.title("Correlation Heatmap (Including Categorical Variables)")
plt.show()
plt.savefig("corr")
```

```python
# Calculate correlations with the output variable
output_var = 'class_live' 
correlations_with_output = correlation_matrix[output_var].abs()

# Select features with an absolute correlation above 0.5
high_corr_features = correlations_with_output[correlations_with_output > 0.5].index
X_Age_high_corr = X_Age_encoded[high_corr_features]

# Display the selected features
print("Selected Features with |correlation| > 0.5 with the output variable:")
print(high_corr_features)
```

```python
# Calculate correlations with the 'malaise' feature
corr_with_malaise = correlation_matrix['malaise'].abs()

# Find features that are highly correlated with 'malaise' (threshold = 0.5)
high_corr_with_malaise = corr_with_malaise[corr_with_malaise > 0.5].index

# Drop these features from the DataFrame
X_Age_final = X_Age_encoded.drop(columns=high_corr_with_malaise)

# Display the remaining features
print("Remaining Features after dropping those highly correlated with 'malaise':")
print(X_Age_final.columns)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Assuming you have your DataFrame as X_Age_encoded with 'class_live' as the target variable

# Hold-out strategy
X_train, X_test, y_train, y_test = train_test_split(
    X_Age_encoded.drop(columns='class_live'),
    X_Age_encoded['class_live'],
    test_size=0.2,  # 20% for testing
    random_state=42  # For reproducibility
)

# Displaying the number of samples in each class for the hold-out strategy
print("Hold-out Strategy:")
print("Training set class distribution:")
print(y_train.value_counts())
print("\nTest set class distribution:")
print(y_test.value_counts())

# Stratified hold-out strategy
X_train_stratified, X_test_stratified, y_train_stratified, y_test_stratified = train_test_split(
    X_Age_encoded.drop(columns='class_live'),
    X_Age_encoded['class_live'],
    test_size=0.2,  # 20% for testing
    random_state=42,  # For reproducibility
    stratify=X_Age_encoded['class_live']  # Stratify by the target variable
)

# Displaying the number of samples in each class for the stratified hold-out strategy
print("\nStratified Hold-out Strategy:")
print("Training set class distribution:")
print(y_train_stratified.value_counts())
print("\nTest set class distribution:")
print(y_test_stratified.value_counts())
```
