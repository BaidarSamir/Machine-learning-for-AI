```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import SGDRegressor
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import seaborn as sns

dataset = pd.read_csv('student_scores.csv')
X=dataset.iloc[:, :-1].values
y=dataset.iloc[:,-1].values
print(X,y)
dataset.head(5) 


```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
# intercept = model.intercept_
# slope = model.coef_[0]
# print(f"Model equation: y = {intercept:.2f} + {slope:.2f} * X")
```

```python
plt.scatter(X, y, color='blue', label='Data points')
plt.plot(X, model.predict(X), color='red', label='Regression line')
plt.title('Hours vs Scores with Regression Line')
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

y_train_pred = model.predict(x_train)
y_test_pred = model.predict(x_test)

# Training set metrics
mae_train = mean_absolute_error(y_train, y_train_pred)
mse_train = mean_squared_error(y_train, y_train_pred)
rmse_train = np.sqrt(mse_train)
r2_train = r2_score(y_train, y_train_pred)

# Test set metrics
mae_test = mean_absolute_error(y_test, y_test_pred)
mse_test = mean_squared_error(y_test, y_test_pred)
rmse_test = np.sqrt(mse_test)
r2_test = r2_score(y_test, y_test_pred)

print("Training Set Performance:")
print("MAE:", mae_train)
print("MSE:", mse_train)
print("RMSE:", rmse_train)
print("R-2:", r2_train)

print("\nTesting Set Performance:")
print("MAE:", mae_test)
print("MSE:", mse_test)
print("RMSE:", rmse_test)
print("R-2:", r2_test)

```

```python
plt.scatter(x_test, y_test, color='blue', label='actual')
plt.scatter(x_test, y_test_pred, color='red', marker='*', label='predicted')
plt.title('actual vs predicted Scores')
plt.xlabel('hours')
plt.ylabel('scores')
plt.legend()
plt.show()

```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import SGDRegressor
from sklearn.impute import SimpleImputer
data = pd.read_csv("HousingData.csv")
data.head(5)
```

```python
correlation_matrix = data.corr()
print("Correlation Matrix:")
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True,cmap='summer',fmt='.2f',linewidths=2,vmax=1,vmin=-1)


```

```python
threshold = 0.6
high_corr_features = correlation_matrix.index[correlation_matrix['MEDV'].abs() > threshold].tolist()
print(f"Highly correlated features with MEDV (threshold = {threshold}):")
print(high_corr_features)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Extract two features with high correlation for a 3D plot
if len(high_corr_features) > 2:
    x_feature = high_corr_features[0]
    y_feature = high_corr_features[1]
    z_feature = 'MEDV'
    
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.scatter(data[x_feature], data[y_feature], data[z_feature], color='blue',alpha=0.3)
    ax.set_xlabel(x_feature)
    ax.set_ylabel(y_feature)
    ax.set_zlabel('MEDV')
    plt.show()
else:
    print("Not enough highly correlated features for a 3D plot.")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import MinMaxScaler

# Load the dataset
data = pd.read_csv("HousingData.csv")

# Check for missing values in the dataset
print("Missing values in the dataset:")
print(data.isnull().sum())

# Select features and target variable
selected_data = data[high_corr_features]
X = selected_data.drop('MEDV', axis=1)
y = selected_data['MEDV']

# Handle missing values using SimpleImputer before splitting the data
imputer = SimpleImputer(strategy='mean')
X_imputed = imputer.fit_transform(X)  # Fit and transform the entire dataset

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_imputed, y, test_size=0.2, random_state=42)

# Normalize the data
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("Data has been normalized using Min-Max Scaler.")
print("Train scaled shape:", X_train_scaled.shape)
print("Test scaled shape:", X_test_scaled.shape)
```

```python
print("test scaled",X_test_scaled)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import SGDRegressor
from sklearn.metrics import mean_squared_error

# Learning rates to test
learning_rates = [0.01, 0.001, 0.0001]

# Loop over each learning rate
for rate in learning_rates:
    # Initialize the SGDRegressor with the current learning rate
    sgdr = SGDRegressor(learning_rate='invscaling', eta0=rate, max_iter=20, verbose=0)

    # List to store loss values (MSE)
    loss_values = []

    # Fit the model for a specified number of iterations (e.g., 100)
    for _ in range(sgdr.max_iter):  # Change this to adjust the number of iterations
        sgdr.fit(X_train_scaled, y_train)
        # You can calculate the current loss (mean squared error) based on predictions
        y_pred = sgdr.predict(X_train_scaled)
        mse = mean_squared_error(y_train, y_pred)
        loss_values.append(mse)

    # Plot the convergence graph
    plt.plot(loss_values, label=f'Learning rate: {rate}')

# Customize the plot
plt.title('Convergence Graph for Different Learning Rates')
plt.xlabel('Iterations')
plt.ylabel('Loss (Mean Squared Error)')
plt.legend()
plt.grid()
plt.show()
```
