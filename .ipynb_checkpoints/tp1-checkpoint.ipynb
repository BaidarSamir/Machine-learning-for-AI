```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

plt.style.use("fivethirtyeight")



xticks_used=np.arange(0,11)
years = np.arange(1, 11, 1)  
yticks_used=np.linspace(2000,4500,11)


data=pd.read_csv('employee_salaries_v2.csv')
year=data["Years"]
salaire=data["Salary"]
X = data['Years'].values.reshape(-1, 1)
y = data['Salary'].values

model = LinearRegression()
model.fit(X, y)
plt.plot(X, model.predict(X), color='red', linewidth=2, label='Regression line')

plt.title("Salaire de chaque employee")
plt.scatter(year,salaire, label="scatter plot of the salaries",alpha=0.7,s=100)
plt.xticks(xticks_used)
plt.yticks(yticks_used)
plt.xlabel("years of experience")
plt.ylabel("salary")
plt.legend()
plt.show()




```
